#### Create new columns for State and Pincode (Seperate state and pincode from address)

```python
df.head()
```

```python
df['seller_address'][0]
```

```python
for i in df['seller_address'][0:10]:
    print(i)
```

```python
df['seller_address'][0].split(' ')[-1]
```

```python
pincode = []
states = []
for i in df['seller_address']:
    if i != 'Not Mentioned':
        
        pincode.append(i.split(' ')[-1])
        states.append(i.split(',')[-1].split('-')[0].strip())
    else:
        pincode.append(np.nan)
        states.append(np.nan)
        
df['pincode'] = pincode
df['states'] = states
```

```python
df.head()
```

#### orders from each state

```python
df['states'].fillna('nan',inplace=True)
```

```python
df['states'] = df['states'].str.lower()
```

```python
df['states'] = df['states'].replace('nan',np.nan)
```

```python
df.groupby('states').size().reset_index(name='count').sort_values('count',ascending=False).head(10)
```
